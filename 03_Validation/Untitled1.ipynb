A feature 'sex' eh do tipo caractere. Precisamos converter os dados dessa coluna para o tipo int

```python
def conv(sex):
    if sex == 'M':
        sex = 0
    elif sex == 'F':
        sex = 1
    else:
        sex = 2
        
    return sex
```

```python
data['sex'] = data['sex'].apply(conv)
data
```
